### 2nd Pattern : Standardize to dd-mm-yyyy Format
the other format is still in MM / dd / yyyy format, we will also standardize it to be dd-MM-yyyy

In the first step, we will separate the date text into three columns, namely day, month and year - where the separator is a slash (/).


```r
sql <- "SELECT tanggal_lahir 
FROM dqlab_messy_data 
WHERE tanggal_lahir like '%/%'"
```

```r
data.pelanggan2 <- read.xlsx("C:\\Users\\aftermath\\Documents\\CS-101\\R\\asset\\data.pelanggan2.xlsx")
data.pelanggan2$tanggal_lahir
```


The colsplit function of the reshape2 package is perfect for achieving this goal. The syntax of colsplit is as follows.
<pre>

</pre>
Where:
<br>string: is the text to be separated.
<br>pattern: regex pattern used to separate text.

The second step is to reassemble the three columns in the order of days, months and years with the paste function and the minus sign (-) as the merge character. The syntax is as follows.
<pre>
         paste (..., sep = "")
</pre>
Where:
<br>...: is a list of variables or comma-separated text.
<br>sep = "": is the intermediate character when merging text. In this example the space character.

```r
data.pelanggan2$tanggal_lahir <- paste(tanggal.split$hari, tanggal.split$bulan, tanggal.split$tahun, sep="-")
data.pelanggan2$tanggal_lahir 
```

### Identifying Non-logical Birth Date Not Logical
The next stage relates to business processes. Let's say our business has only been established for 5 years and only received a maximum of 75 years. Thus the maximum age of the registered customer must be 80 years (75 + 5). If more than that, then it is assumed that the wrong note.

To calculate age, we actually calculate the difference between the date of birth and the date of reference that we use. This goal can be found by using the difftime function.

Here is an example of its use.
<pre>
    difftime(tanggal_referensi , data.pelanggan$tanggal_lahir , units = "days")
</pre>

Where:
<br>difftime: is a function to calculate the difference between two dates.
<br>tanggal_referensi: variable with Date data type that will be a reference to calculate age.
<br>pelanggan$tanggal_lahir: variable with Date data type that will be compared.
<br>unit = "days": the difference in units of days (days). In addition to days we can enter seconds (secs), minutes (mins), hours (hours), and weeks (weeks).


Note, because there is no year - then we need to convert this difftime into years using the as.Numeric function and then divide by 365. Complete it as follows.

<pre>
    as.numeric(difftime(tanggal_referensi , data.pelanggan$tanggal_lahir , units = "days"))/365
</pre>

Where:
<Br>numeric: is a function to convert data to numeric.
<br>/ 365: divide the number by 365.

```r
#Set tanggal referensi ke 27 April 2018
tanggal_referensi <- as.Date("27-4-2018", "%d-%m-%Y")
tanggal_referensi
```

```r
#tanggal_lahir as.Date
data.gabungan$tanggal_lahir <- as.Date(data.gabungan$tanggal_lahir, "%d-%m-%Y")

#Calculates the difference in dates in years
data.gabungan$umur <- as.numeric(difftime(tanggal_referensi, data.gabungan$tanggal_lahir,units = "days"))/365

#Check the maximum age of 80 years
data.gabungan$umur_valid <- data.gabungan$umur <= 80 & data.gabungan$umur > 0

#Reformat birthday
data.gabungan$tanggal_lahir <- format(data.gabungan$tanggal_lahir, "%d-%m-%Y")

#Show data
data.gabungan
```
