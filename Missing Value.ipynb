### NA


NA is the main representation of the missing value in R, and is an atomic value.
NA positions are the same as examples of atomic values of numbers such as 1, 2 or 100. Also the same as examples of character values such as "a", "b", or "c".

Because numbers and characters like the example above are often referred to as constants, then NA is a constant for missing values.

Thus, briefly NA is:
-Replace missing values
-It is a constant or an atomic value

NA cannot be used as a variable because it is a keyword.

NA = both characters are uppercase.

```r
NA
```

```r
typeof(NA)
```

```r
x <- NA
print(x == NA)
print(is.na(NA))
```

```r
typeof(NA_integer_)
typeof(NA_real_)
typeof(NA_complex_)
typeof(NA_character_)
is.na(NA_integer_)
is.na(NA_real_)
is.na(NA_complex_)
is.na(NA_character_)
```

```r
#Create a vector named container with the contents of numbers, where one of them has a missing value
container <- c(1,2,3,NA,3,1)

#Check the entire data type with lapply and typeof loop
lapply(container, typeof)

#Use is.na to check for the missing value of each element in the vector
is.na(container)
```

### NULL
NULL is a value that is widely used in other programming languages or SQL (Structured Query Language) to represent objects that are missing or null objects (missing objects) or missing values.

In R, the missing value is represented by NA. Now, for this missing object we use NULL in R.

NULL is a special object in R that is used to represent an undefined value or object.

NULL represents an undefined object and is not considered by a vector. It is thus not part of the vector.

```r
#NULL with Vector
#Create a vector named container with the contents of numbers, where one of them has a missing value and one of them is a NULL
container <- c(1, 2, 3, NA, 5, NULL, 7)

#Show the vector, NULL is removed
print(container)

#check if removed
length(container)
```

```r
#NULL with List
#Create a list named container with the contents of numbers, where one of them has a missing value and one of them is a NULL
list <- list(1, 2, 3, NA, 5, NULL, 7)

#Show the list, NULL is removed
print(list)

#check if removed
length(list)
```
