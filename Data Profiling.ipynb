# Data Profiling
Data profiling is the earliest stage to perform data cleansing. In the process of this activity that is simple but important:
<br>-We will identify patterns contained in a column of data.
<br>-And compare it with reasonable scientific expectations or measurements, to find data that needs to be improved.
Profiling technique can be done in many ways, but certainly in general will search the entire data.

Because patterns can be of many kinds, we will focus profiling on data content with simple but quite effective text patterns.

In order to be practiced in real terms, we will specifically use the following functions and operators throughout this chapter.

Function summary of the default package R.
Function library basic_pattern_analysis of bpa in R.
Using the == operator and function grepl to pull the data for patterns of findings.

```r
#Required Data
library(openxlsx)
library(bpa)
data.pelanggan <- read.xlsx("C:/Users/aftermath/Documents/CS-101/R/asset/dqlab_messy_data_pelanggan.xlsx", sheet="Pelanggan")
data.pelanggan
```

### Summary
Function summary is a function that will provide a brief summary of the data by analyzing the contents of the data.
<br>For numeric data types, the summary will provide minimum, maximum, median, mean, and others.
<br>For character data types will report only the data type and length.
<br>For the factor data type will contain the factor values and the number of occurrences of the data (frequency).

```r
summary(data.pelanggan)
```

### basic_pattern_analysis function
The output of this function is character recognition per character into the following symbol:

<br>Each uppercase A to Z will be represented by the letter A.
<br>Each lowercase letter a / z will be represented by a letter.
<br>Each number from 0 to 9 will be represented by number 9.
<br>Spaces and tabs will be represented by the letter w.
<br>All symbols will be represented by themselves. Example: a minus sign (-) will still be represented by a minus sign (-).
<br>Missing value NA will be represented by NA.
<br>NaN (Not a Number) will be represented as "AaA".

```r
basic_pattern_analysis(x="DQLab")
basic_pattern_analysis(x="17 Agustus 1945")
basic_pattern_analysis(x=3.14)
```

### Profiling of Vectors

```r
basic_pattern_analysis(c("KD-001", "DQLab", "KD-002"))
```

### Profiling with unique_only=TRUE

```r
basic_pattern_analysis(c("KD-008", "012345", "KD-010"), unique_only=TRUE)
```

### Outlier Detecting - Kode Pelanggan

```r
basic_pattern_analysis(data.pelanggan$Kode.Pelanggan, unique_only = TRUE)
```

```r
basic_pattern_analysis(data.pelanggan$Kode.Pelanggan)=="AA-9999"
```

```r
data.pelanggan[ basic_pattern_analysis(data.pelanggan$Kode.Pelanggan)=="AA-9999" , ]
```

### Outlier Detecting - Nama
We can simply say an unusual name if:
<br>Contains non-letter characters, spaces, periods and commas. -> [^Aaw.,]
<br>Having more than one space in sequence. -> ww

```r
data.pelanggan[grepl(pattern="[^Aaw.,]", x=basic_pattern_analysis(data.pelanggan$Nama)),]

data.pelanggan[grepl(pattern="ww", x=basic_pattern_analysis(data.pelanggan$Nama)),]
```

### Combining the Results of Profiling into the Initial Dataset
The three steps of the merging process are as follows:
<br>-Profiling all columns of customer data and saving it to a new variable, in the following example the name is customer.data.data pattern.
<br>pattern.data.customers <- basic_pattern_analysis (data.customers)
<br>-Change the names of the columns in the customer's variable.data.ata variable by adding the "Pattern" prefix. with the following command.
<br>names (pattern.data.customer) <- paste ("Pattern", names (pattern.data.customer), sep = ".")
<br>-Combine both customer data.frame data and customer.data.data with the cbind function, and save it back to customer data. variable.
<br>customer data <- cbind (customer data, customer data pattern)

```r
#Profiling all columns of customer data and saving it to a new variable, in the following example the name is customer.data.data pattern.
pola.data.pelanggan <- basic_pattern_analysis(data.pelanggan)

#Change the names of the columns in the customer's variable.data.ata variable by adding the "Pattern" prefix. with the following command.
names(pola.data.pelanggan)<-paste("Pola",names(pola.data.pelanggan),sep=".")

#Combine both customer data.frame data and customer.data.data with the cbind function, and save it back to customer data. variable.
data.pelanggan <- cbind(data.pelanggan, pola.data.pelanggan)

str(data.pelanggan)
```
