# Principal Component Analysis : Dimension Reduction
the large number of variables raises its own problems, namely the large number of redundant variables, the difficulty in visualizing, the difficulty in explaining the model and the amount of storage required. This is where the Principal Component Analysis technique can play a role, namely as an unsupervised algorithm to reduce the number of variables (dimension reduction) to be used as input for other algorithms without greatly reducing the quality of the prediction rating.

### Principal Component Analysis
Principal Component Analysis (PCA) is a method of dimension reduction in machine learning. PCA will choose "variables" that can explain most of the variability of the data. If you are familiar with Regression statistical analysis, the concept of variability is similar to the coefficient of determination R2.
PCA reduces dimensions by forming new variables called Principal Components which are linear combinations of old variables.

This Principal Component is selected with the following conditions:
<br>-each Principal Component provides the greatest variance,
<br>-each Principal Component has no correlation with other PCs and
<br>-the length of the vector containing the coefficients of this PC is 1.

The basic concept of PCA is to approve by looking for the variables that provide the greatest variability in the data and then place the "axes" of principal components so that the relationships between variables (correlation between variables) are minimal. Calculation of Variance and Principal Components can be done using the concept of eigenvalue (eigenvalue) and eigenvector (eigenvector) from Linear Algebra. Eigenvalues indicate the contribution of a principal component to the data variance. Eigenvectors provide Principal Component coefficients. The number of Principal Components to be selected is adjusted to the desired limit or by using Screeplot. How to calculate eigenvalues and how to obtain eigenvectors are discussed in the review chapter in Linear Algebra.

### Assumptions in PCA
Assumptions that must be met for the Principal Component Analysis technique to work properly:
<BR>-Variable types are numeric.
<BR>-The variables that become PCA input are predictor variables.
<BR>-Predictor variables have a linear relationship.
<BR>-All variables must be standardized (centered and scaled).

### PCA Steps
Following are the steps needed to conduct Principal Component Analysis:
<br>-Prepare data (data standardization),
<br>-Calculating covariance matrices or correlation matrices,
<br>-Calculate eigenvalues and eigenvectors from correlation matrices,
<br>-Choose the principal component,
<br>-Output visualization,
<br>-Calculate new scores.

How the Principal Component Analysis technique works is to choose the principal component that provides the greatest variance. The workings of the PCA method cause this method to be very sensitive to the values of different variables or to the scale of different variables. So that PCA does not "pick" the wrong variables, the input variables need to be standardized. Standardization of variables is done by reducing each observation on the variable with the mean of the variable and dividing by the standard deviation of the variable.

The correlation coefficient shows the strong linear relationship between 2 variables. The relationship between the 2 variables for all predictor variables in the data can be seen in the correlation matrix. A high correlation coefficient (close to 1 for direct linear relations or close to -1 for opposite directions) is an indication of redundancy. This is where the PCA technique can be used to reduce dimensions by selecting new "uncorrelated" variables that can explain most of the variability of data.

From the correlation matrix obtained information about the strength of the linear relationship between variables. Our goal in conducting PCA is to choose the variables that provide the greatest variance by minimizing distortion between variables. This is done by calculating the eigenvalues and eigenvectors from the correlation matrix. The "new variable" principal component is obtained from the eigenvector and is a linear combination of the old variable. The coefficients indicate the magnitude of weight of the old variables in each principal component.

There are several criteria that can be used to determine k:
<br>-Limitation of variability, for example 80%.
<br>-Kaiser Kriterion: selects all PCs whose eigenvalues are greater than 1.
<br>-Select k Principal Component.

In this module we will choose the first Principal Component whose eigenvalues are more than 1 up to certain variability limits. Why is the eigenvalue selected more than 1 eigenvalue? A PC with an eigenvalue of more than 1 is able to explain greater data variability than the original variables.

The amount of contribution of a principal component to the variability of data can be calculated using the formula
<pre>
λj / p
</pre>

where λj is the eigenvalue to j and p is the total number of variables.

Visualization of PCA output can be done using biplots. This plot displays the scores and loading in the same graph using the 2 most important Principal Components. The horizontal axis displays the first PC, the vertical axis displays the second PC. Variables that have a large contribution to PC1 will be drawn horizontally and variables that have a high contribution to PC2 will be drawn vertically.

Old variable scores will be transformed into new variable scores using formulas:
<pre>
newdata = olddata * PC


### Limitations of PCA
In practice Principal Components are given their own names based on a summary of the dominant variables. This requires domain knowledge and there are times when the variables that appear are not in accordance with the theory in the domain.
The way PCA works in reducing dimensions is to form Principal Components that provide the greatest data variability. The limitation of PCA is that the PC selection process is only carried out with predictor variables. Therefore PCA should not be used as a model but rather be used as a data preprocessing technique to then be input to other methods. Another alternative is to use the Partial Least Squares method which involves response variables in dimension reduction.

R provides the prcomp () and princomp () functions to perform Principal Component Analysis.
<PRE>
prcomp (x, center = TRUE, scale. = FALSE)
</PRE>
Description
<PRE>
The argument x is a dataframe containing numeric or matrix variables.
R automatically standardizes the data when the center = TRUE and scale = TRUE arguments are activated.
Stdev output is the root of the eigenvalue.
The output center produces the mean of the values ​​of a variable (the mean column in the data structure used in the example).
The output scale produces the standard deviation for the values ​​of a variable (the column mean in the data structure) used in the example).

Output rotation displays the Principal Component loading which will be used to calculate data scores on the new coordinate system. The coefficients in the Principal Component vectors indicate the magnitude of the correlation between the variable and the Principal Component.
Output x displays the data score on the "new coordinate system".

The "new variables" principal components PC1, PC2, and PC3 are obtained from eigenvectors and are linear combinations of the old variables x1, x2 and x3. These coefficients indicate the contribution of the old variables in each principal component.

The number of PCs (p) chosen is when there is a sharp decrease in variance before the decrease in the sloping variance. In this example the number of PCs chosen is 2. If we use the Kaiser criterion in the scree plot, then the number of PCs chosen is 1.

### 4 Variables Dataset | Split Training and Testing Data
Data will be separated into 2 parts: training and testing set. In order for the training and testing set to obtain data with all risk ratings, sampling will be conducted using the stratified sampling method. 80% of the data from each risk rating will be allocated for the training set, the rest will be allocated for the testing set.

To find the row index that has a risk rating of 1, 2 and so on, use the which () function.

```r
index1 <- which(dat_raw$riskrating == 1)
index2 <- which(dat_raw$riskrating == 2)
index3 <- which(dat_raw$riskrating == 3)
index4 <- which(dat_raw$riskrating == 4)
index5 <- which(dat_raw$riskrating == 5)
```

The amount of data for each risk rating allocated for the training set is calculated by the following command.

```r
train_pct <- 0.8
ntrain1 <- round(train_pct * length(index1))
ntrain2 <- round(train_pct * length(index2))
ntrain3 <- round(train_pct * length(index3))
ntrain4 <- round(train_pct * length(index4))
ntrain5 <- round(train_pct * length(index5))
```

Specify the seed so that sampling can be replicated with set.seed (). Take a sample of each risk rating data to be allocated into the training set.

```r
set.seed(200)
train1_index <- sample(index1, ntrain1)
train2_index <- sample(index2, ntrain2)
train3_index <- sample(index3, ntrain3)
train4_index <- sample(index4, ntrain4)
train5_index <- sample(index5, ntrain5)
```

Use the setdiff () function to allocate data for the testing set. The selected row index is the row index that was not taken in the sampling process for training set

```r
test1_index <- setdiff(index1, train1_index)
test2_index <- setdiff(index2, train2_index)
test3_index <- setdiff(index1, train3_index)
test4_index <- setdiff(index1, train4_index)
test5_index <- setdiff(index1, train5_index)
```

Combine the sampling results of each rating into the training set.

```r
csdat_train <- do.call("rbind", list(dat_raw[train1_index,],
dat_raw[train2_index,], dat_raw[train3_index,],
dat_raw[train4_index,], dat_raw[train5_index,]))
cs_train <- subset(csdat_train, select = -riskrating)

head(cs_train)
```

Combine the sampling results of each rating into the training set.

```r
csdat_test <- do.call("rbind", list(dat_raw[test1_index,],
dat_raw[test2_index,], dat_raw[test3_index,],
dat_raw[test4_index,], dat_raw[test5_index,]))
cs_test <- subset(csdat_test, select = -riskrating)

head(cs_test)
```
