### Database Connection
Description of the parameters used:
<Br>-MySQL (): is a function that is a must to load the MySQL driver to R.
<Br>-host: this is the location of the server, it can be in the form of an IP address or hostname. To practice this course, we use the host name "mysqlhost".
<Br>-user: this is the name of the user that is allowed to connect to the server. To practice this course, we use user c.
<Br>-password: the password used by the user. To practice this course, we use the "demo" password.
<Br>-dbname: name of the database used. To practice this course, we use a database named "dqlabdatawrangling".

```r
library(RMySQL)

#Open Connection
con <- dbConnect(MySQL(), user="demo", password="demo", host="mysqlhost",
                 dbname="dqlabdatawrangling")
#Close Connection
all_cons <- dbListConnections(MySQL())
for(con in all_cons) dbDisconnect(con)
```

### Acquire Data using Select

```r
#Construct SQL
sql <- "SELECT * FROM dqlab_messy_data"

#Send query to server
response <- tryCatch(dbSendQuery(con, sql), finally = print("query ok"))

#Acquire Data
data.pelanggan <- fetch(response, n=-1)
str(data.pelanggan)
```

If we want to find data that has commas, then we can write the like construction as follows.

Nama like '%,%'

This means that the comma name is enclosed in any character - the percent sign to the left and right of the comma.

Then if we want to find a name beginning with the letter K, then we can use the following construction like.

Nama like 'K%'

For example : SELECT kode_pelanggan, nama FROM dqlab_messy_data WHERE nama like 'a%'

This filter construction will search for all data starting with a. Upper and lower case letters are not a problem.
