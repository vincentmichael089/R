### Introduction to the Hypothesis Test
Hypothesis testing is a decision-making method based on data analysis. In statistics, you can test a hypothesis as true or false. There are 2 types of hypotheses, the null hypothesis (null hypothesis) and the alternative hypothesis. The null hypothesis (Ho) is a hypothesis that is contrary to the theory to be proven. Alternative hypothesis (Ha) is a hypothesis related to the theory that will be proven.

In conducting statistical tests we need to use the statistical test method, which is often used, namely z-test, t-test, chi-square test, and f-test. In this chapter we will not discuss the details of each test statistic above, but we will focus on how to use it.

Next we have to understand the p-values ​​and alpha that will be used in test statistics.

P-value is the smallest opportunity to reject Ho. Whereas alpha is the error rate. Alpha values ​​are usually 1%, 5%, and 10%. In practice alpha 5% is often used, because it is more moderate.

Hypothesis H0 is rejected if the p-value is less than alpha (5%), whereas if the p-value is more than alpha then H0 is accepted.

### Difference between Descriptive Statistics and Statistics of Inference
In statistics there are 2 types of data analysis, namely descriptive statistics and inference statistics. Hypothesis testing, which was explained in the previous section included in the statistics of inference.

To distinguish between the two types of analysis above, one can listen to the following explanation:

<br>Descriptive Statistics are statistics used for data analysis by drawing sample data without making conclusions for population data. Some things that can be done are the presentation of data through tables, graphs, calculation mode, median, mean, calculation of data spread through calculation of averages and standard deviations. Descriptive statistics are used for data exploration.

<br>Inference statistics are the statistics used to analyze sample data and the results are applied to the population. Some things that can be done are to test hypotheses with test statistics, such as chi-square test, student-t test, f-test, z-score test.
Inference statistics can be used to confirm the results of descriptive statistics.


## Relationship Between Numeric Variables
There are two ways to see the relationship between variables, namely the scatter plot graph and correlation analysis. Scatter plot graph to see the direction of the relationship, positive and negative. While the correlation analysis is to test / confirm whether the two variables are indeed related and how strong the relationship is.

The correlation coefficient ranges from -1 to 1. Strong correlation when approaching -1 or 1, while said to be weak if approaching 0. To find out whether there is a relationship or not using correlation analysis, with the following hypothesis
<br>Hypothesis null (null): there is no relationship between the two variables.
<br>Alternative hypothesis: there is a relationship between the two variables.

### Scatter Plot

```r
plot(data_intro$Pendapatan,data_intro$Total)
```

### Correlation Testing
In the correlation test to test it we use a t-test. With the following hypothesis:
<br>H0: there is no relationship between income and total expenditure.
<br>Ha: there is a relationship between income and total expenditure

```r
cor.test(data_intro$Pendapatan,data_intro$Total)
```

## Relationship Between Categorical Variables
The relationship between the two can be seen using cross tabulation and can also be seen trends. In the relationship between categorical and categorical variables it cannot be known how strong the relationship between the two is and how they are affected (positive or negative). To find out whether or not there is a relationship using the chi-square test statistical test, with the following hypothesis:

Null hypothesis: there is no relationship between the two variables<br>
Alternative Alternative Hypothesis: there is a relationship between the two variables

### Chi-Square
The table command is for viewing tabulations between categorical variables, while the chisq.test command is used to view relationships statistically.

With the following hypothesis:

<br>H0: there is no relationship between product type and level of satisfaction.
<br>Ha: there is a relationship between product types and satisfaction levels

```r
table(data_intro$Produk,data_intro$Tingkat.Kepuasan)
chisq.test(table(data_intro$Produk,data_intro$Tingkat.Kepuasan))
```

## Relationship Between Categorical Variables and Numerical Variables
The relationship between the two can be seen by comparing the averages in each category. If the average value is different then the two variables have a relationship. In the relationship between categorical and numerical variables it cannot be known how strong the relationship is between the two and how they are affected (positive or negative).

To find out whether or not there is a relationship using the t-test statistical test, with the following hypothesis

<Br>Null hypothesis: there is no relationship between the two variables
<Br>Alternative Alternative Hypothesis: there is a relationship between the two variables

### t.test
The t-test hypothesis is as follows:

<Br>Null hypothesis: there is no difference in average total spending between men and women
<br>Alternative hypothesis: there is a difference in average total expenditure between men and women

```r
#the relationship between categorical and numeric data, i.e. gender and total expenditure variables
boxplot(Total~Jenis.Kelamin, data = data_intro)

t.test(Total~Jenis.Kelamin,data = data_intro)
```
