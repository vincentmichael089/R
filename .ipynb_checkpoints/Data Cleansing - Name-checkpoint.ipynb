# Cleansing : Name

### Remove Duplicate "Space" between Words

```r
name <- " jupyter  notebook "
name <- gsub( " {2,}", " ", name) #remove doubled-to-infinity space
name
```

### Remove "Space" at Start and End of Words

```r
name <- " jupyter  notebook "
name <- trimws(name, which = "both") #remove space at start and end of words with "both"
name
```

### Remove Numbers and Symbols

```r
name <- " jupyte3$r  notebook123 "
name <- gsub("[^a-zA-Z .,]", "", name) #[^ A-Za-z.,]: Regex patterns not letters, spaces, periods and commas.
name
```

### Normalize Titles

```r
name <- "Bapak Jupyter Notebook"
name <- gsub("\\bBapak\\b", "", name, ignore.case = TRUE)
name
```

```r
name <- "IR. Jupyter Notebook"
name <- gsub("\\bir\\b", "Ir.", name, ignore.case = TRUE)
name
```
