```r
library(RMySQL)
#Membuka koneksi
con <- dbConnect(MySQL(), user="demo", password="demo", host="mysqlhost",
                 dbname="dqlabdatawrangling")

#Konstruksi SQL
sql <- "SELECT pola_no_telepon, length(pola_no_telepon) as panjang_text, count(*) as jumlah_data
from dqlab_messy_data
group by pola_no_telepon"

#Mengirimkan query
rs <- tryCatch(dbSendQuery(con, sql), finally = print("query ok"))

#Mengambil data
profil_no_telepon <- fetch(rs, n=-1)
print(profil_no_telepon)

#Clear resultset
dbClearResult(rs)

#Menutup Koneksi
all_cons <- dbListConnections(MySQL())
for(con in all_cons) dbDisconnect(con)
```

The following is an analysis of the results above.
<BR>1. There are seven patterns identified.
<BR>2. The majority of the numbers are in pattern number 2, 6 and 7.
<BR>3. The length of the data at 2, 6 and 7 are 17, 15 and 16. But because of pattern no 2 there is a + sign, so the actual length is probably 16. The same as pattern no 7.
<BR>4. There are no other characters besides the numbers and + signs. Except for one data in pattern no. 3 is just a minus sign (-). This can be considered a missing value.
<BR>5. Other unusual patterns are pattern no 1 (+9999999999999999, there are only 2 data), 4 (9999999999999, only 1 data), and 5 (9999999999999999, only 3 data).
<BR>6. Patterns 1, 3, 4 and 5 will be flagged as outliers that need to be followed up in further practice.
<BR>7. There is a difference in the number of texts between patterns no. 5, 6 and 2 (the number of texts is 15, 16 and 17, respectively), we assume this is due to the initial mobile numbers 0, 62 and +62 in front of it. We will check this with additional queries in the next profiling.

### Standardize Phone Column
Assume the standard phone number is  +62xxxxx

```r
library(openxlsx)
library(bpa)
data.telepon <- read.xlsx("C:/Users/aftermath/Documents/CS-101/R/asset/no_telepon.xlsx")
data.telepon
```

```r
#add + to 9999999999999999 (62xxxxx) pattern using paste function
data.telepon[data.telepon$pola_no_telepon=="9999999999999999",]$no_telepon <- paste("+", data.telepon[data.telepon$pola_no_telepon=="9999999999999999",]$no_telepon, sep="")
```

```r
#change (08xxxxx) to (+62xxxxx) using gsub function
data.telepon[data.telepon$pola_no_telepon=="999999999999999",]$no_telepon <- gsub("^0","+62",data.telepon[data.telepon$pola_no_telepon=="999999999999999",]$no_telepon)
```

```r
#check pattern
data.telepon$pola_no_telepon <- basic_pattern_analysis(data.telepon$no_telepon)
data.telepon
```

```r
#Summary
data.telepon$pola_no_telepon <- as.factor(data.telepon$pola_no_telepon)
summary(data.telepon$pola_no_telepon)
```
