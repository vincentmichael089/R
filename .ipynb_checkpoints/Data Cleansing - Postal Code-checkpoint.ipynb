You can see the letter O, which should be 0 (zero) for customer_code = 'KD-00093' and for customer_code = 'KD-00085'. 
<br>After that there is the letter I which probably means the number 1.

In the next stage we try to replace it first with our assumptions, O becomes 0. And I becomes 1.

### Standardize Postal Code

```r
library(openxlsx)
data_kode_pos <- read.xlsx("C:/Users/aftermath/Documents/CS-101/R/asset/kode_pos.xlsx")
data_kode_pos
```

```r
data_kode_pos$kode_pos <-gsub("O","0",data_kode_pos$kode_pos)
data_kode_pos$kode_pos <-gsub("I","1",data_kode_pos$kode_pos)
data_kode_pos
```

```r
library(bpa)

data_kode_pos$pola_kode_pos <- basic_pattern_analysis(data_kode_pos$kode_pos)
data_kode_pos
```
