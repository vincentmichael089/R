# Fundamental R

### Variabel

```r
pi <- 3.14
print(pi)
```

### Vector

Vector menyimpan 1 jenis data saja

```r
vec1 <- c(5:10)
print(vec1)
```

```r
vec2 <- c("ayah","ibu")
print(vec2)
print(vec2[1])
print(vec2[2])
```

```r
bio <- c(name = "vincent", job = "student")
print(bio)
print(bio["name"])
```

### List

Vector menyimpan lebih dari 1 jenis data

```r
lis <- list(2, "Budi", 4, TRUE)
print(lis[2])
print(lis[[2]])
```

```r
bio2 <- c(name = "vincent", umur = 21.2)
print(bio2)
print(bio2["umur"])
```

### Data Frame

```r
huruf <- c("satu","dua","tiga","empat")
simbol <- c(1,2,3,4)
dataframe <- data.frame(huruf, simbol)
print(dataframe)
print(dataframe$huruf)
```
