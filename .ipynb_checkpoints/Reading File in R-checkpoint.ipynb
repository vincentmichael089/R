### check.names Function
If we look at the execution of read.csv and its results, we see a column with X prefixes - that is, X, X.1, X.2, and so on. This happens because read.csv has detected an empty and more than column names. This condition will be automatically "fixed" by the read.csv function by adding an X prefix in front of the column.

```r
penduduk.dki <- read.csv("C://Users//aftermath//Documents//CS-101//R//asset//dkikepadatankelurahan2013.csv",sep = ",", check.names = FALSE)
penduduk.dki
```
