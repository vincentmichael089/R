### kmeans Function
Function kmeans require at least 2 parameters, namely:
<pre>
x: the data used, where all the contents of the data must be numeric.
centers: number of clusters desired.
</pre>
Best here means that the distance between each point to the mean of the cluster itself is smaller than the mean of the other cluster.

Note: Mean or average value here is often referred to as centroids in various data science literature.

And this kmeans function is usually accompanied by a seet.seed function call. This is useful so that we "uniform" list the same random values ​​from kmeans so that we get the same output.

Following is an example of using the set.seed and kmean combination functions.
<pre>
set.seed (100)

kmeans (x = customer [c ("Age", "Profession.1")], centers = 3)
</pre>

This means we divide customer data based on "Age" and "Profession" into 3 segments.

Sometimes based on DQLab experience, data parameters and number of segments are not enough. It is necessary to use the third parameter, nstart, which is the number of random combinations generated internally by R. And in the amount that we provide, the algorithm will choose which is the best of those combinations.

```r
set.seed(100)

segmentasi <- kmeans(x = pelanggan[c("Jenis.Kelamin.1", "Umur", "Profesi.1", "Tipe.Residen.1","NilaiBelanjaSetahun")],
    center = 5, nstart = 25)

segmentasi
```

### Analysis of Sum of Squares Results
The concept of sum of squares (SS) is the sum of the "square distances" difference between each data point with its mean or centroid. This SS can be the mean or centroid for each cluster or the whole data. Sum of squares in other data science literature is often called the Sum of Squared Errors (SSE).

The greater the value of SS states the wider the difference between each data point in the cluster.

<pre>
Within cluster sum of squares by cluster:
[1]  58.21123 174.85164 316.73367 171.67372 108.49735
 (between_SS / total_SS =  92.4 %)
</pre>

The following is an explanation for the output results kmeans above:
<br>1. The value 316.73367 is the SS for the 3rd cluster, 58.21123 is the SS for the 1st cluster, and so on. The smaller the value, the better the potential.
<br>2. total_SS: is SS for all points relative to the global average, not per cluster. This value is always fixed and is not affected by the number of clusters.
<br>3. between_SS: is total_SS minus the number of SS values in the whole cluster.
<Br>4. (between_SS / total_SS) is the ratio between between_SS divided by total_SS. The greater the percentage, the better the model.

### Simulation of Number of Clusters and SS
The elbow metric method used as the basis of justification is the Sum of Squares (SS), or rather the tot.withinss component of the kmeans object.

This metric will find progressive tot.withinss values for each combination of the number of clusters, and stored in vector form in R.

For this purpose, we will use sapply. The sapply function will be used to call the kmeans function for a range of cluster numbers. We will use this range from 1 to 10.

```r
sse <- sapply(1:10, function(number_of_clusters){
    kmeans(pelanggan[c("Jenis.Kelamin.1", "Umur", "Profesi.1", "Tipe.Residen.1","NilaiBelanjaSetahun")], 
           number_of_clusters, nstart=25)$tot.withinss
    }
)
```

```r
sse
```

### Elbow Effect Graph

```r
library(ggplot2)

jumlah_cluster_max <- 10

ssdata = data.frame(cluster=c(1:jumlah_cluster_max),sse)

ggplot(ssdata, aes(x=cluster,y=sse)) +
                geom_line(color="red") + geom_point() +
                ylab("Within Cluster Sum of Squares") + xlab("Jumlah Cluster") +
                geom_text(aes(label=format(round(sse, 2), nsmall = 2)),hjust=-0.2, vjust=-0.5) +
  scale_x_discrete(limits=c(1:jumlah_cluster_max))
```
