### Remove Column
use operator %in% to take data from vector after the operator

```r
penduduk.dki.csv <-read.csv("C://Users//aftermath//Documents//CS-101//R//asset//dkikepadatankelurahan2013.csv", sep=",")

penduduk.dki.csv <- penduduk.dki.csv[,!names(penduduk.dki.csv) %in% c("X", "X.1","X.2","X.3","X.4","X.5","X.6","X.7","X.8","X.9","X.10","X.11")]

names(penduduk.dki.csv)
```

### Factorize Column
there is a difference in behavior between read.xlsx and read.csv, i.e. by default, read.xlsx does not convert categorical columns as text.

```r
library(openxlsx)

penduduk.dki.xlsx <-read.xlsx(xlsxFile = "C://Users//aftermath//Documents//CS-101//R//asset//dkikepadatankelurahan2013.xlsx")

#before factorized
str(penduduk.dki.xlsx$NAMA.PROVINSI)

penduduk.dki.xlsx$NAMA.PROVINSI <- as.factor(penduduk.dki.xlsx$NAMA.PROVINSI)

#after factorized
str(penduduk.dki.xlsx$NAMA.PROVINSI)
```

### grep Column Name with Unique Words

```r
penduduk.dki.xlsx <-read.xlsx(xlsxFile = "C://Users//aftermath//Documents//CS-101//R//asset//dkikepadatankelurahan2013.xlsx")

#Show the names of the columns containing the word "woman".
pola_nama_perempuan <- grep(pattern="perempuan", x = names(penduduk.dki.xlsx), ignore.case=TRUE)
names(penduduk.dki.xlsx[pola_nama_perempuan])
```

### rowSums Column Name with Unique Words

```r
penduduk.dki.xlsx <-read.xlsx(xlsxFile = "C://Users//aftermath//Documents//CS-101//R//asset//dkikepadatankelurahan2013.xlsx")

#Show the names of the columns containing the word "woman".
pola_nama_perempuan <- grep(pattern="perempuan", x = names(penduduk.dki.xlsx), ignore.case=TRUE)
names(penduduk.dki.xlsx[pola_nama_perempuan])

penduduk.dki.xlsx$PEREMPUAN35TAHUNKEATAS  <- rowSums(penduduk.dki.xlsx[pola_nama_perempuan])

penduduk.dki.xlsx$PEREMPUAN35TAHUNKEATAS
```
