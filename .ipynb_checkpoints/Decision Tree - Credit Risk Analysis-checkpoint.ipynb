# Credit Risk Analysis
Credit risk is the risk that must be borne by an individual or institution when giving loans - usually in the form of money - to individuals or other parties.

This risk is in the form of non-payment of the principal and interest of the loan, resulting in the following losses:

cash flow disruption so that working capital is disrupted.
increase operational costs to pursue these payments (collection).
To minimize this credit risk, a process is usually called a credit scoring and credit rating of the borrower. The output of this process will be the basis for determining whether new loan applications are accepted or rejected.

###  Credit Score
Credit score is the risk value given to an individual or organization applying for a loan based on the track record of the loan and payment made. The process of granting credit scores is usually referred to as credit scoring.

Credit score calculations are usually made based on historical data on the length of late payment and those who have not paid at all (bad debt). Bad debt usually results in lending institutions having to confiscate assets or write off.
Credit score scores usually vary between institutions. However, many later adopted the FICO Score model, which had a range of values of 300 - 850. The higher the value obtained, the better the level of ability of a person or an institution to repay loans.

### Risk Rating
Sometimes many institutions use risk rating or level of risk. Upside with a credit score, the higher this rating indicates the increasing risk.

In addition, the codification is also made simpler than the range of values so that decisions can be taken more quickly. For example, suppose the use of combinations such as the letters AAA, AA +, P-1, and so on. Or for many internal lending institutions, categorization only uses a small range of numbers, for example 1 to 5.

### Analysis and Model of Decision Making
Still related to the previous data example, but with the whole data example - DQLab will provide an illustration of the follow-up activity to the data with the following sample scenario.

An analyst will search our data for patterns. Here are the findings:
<br>-If the number of dependents is more than 4, the risk tendency is very high (ratings 4 and 5).
<br>-If the loan duration is longer, which is more than 24 months, the risk tendency will also increase (ratings 4 and 5).

From these two findings, the analyst will form rules to guide the decision making (decision making model) of the new loan proposal for the following:
<br>-If the number of dependents is less than 5 people, and the duration of the loan is less than 24 months, the rating is given a value of 2 and the loan application is accepted.
<br>-If the number of dependents is more than 4 people and the loan duration is more than 24 months, the rating is given a value of 5 and the loan application is rejected.
<br>-If the number of dependents is less than 5, and the duration of the loan is less than 36 months, the rating is given a value of 3 and the loan is given.
Well, we refer to the three rules as a model for predicting the value of risk rating and becoming the basis for making decisions on new loan applications.

With this model, lending institutions will make decisions more quickly and with a minimum decision making error rate.

### Decision Tree with C5.0
the decision making process with the series construction "If the condition is x then .... other than that ...." or known as "IF ... ELSE ...".

C5.0 is the code for naming an algorithm for a decision tree. Many other algorithms such as random forest, CART, CHAID, MARS, and others. But C5.0 is a very popular algorithm because it has a very good performance in terms of speed and accuracy.

This algorithm is often categorized as classification, where the aim is to categorize or classify something - in our example risk rating - based on input from other data.


```r
library("C50")

data$risk_rating <- as.factor(data$risk_rating) 
input_columns <- c("durasi_pinjaman_bulan", "jumlah_tanggungan")
data.feed <- data[ , input_columns ]
```

set.seed (100) is a command to uniformly retrieve random numbers across R. applications

sample (900, 800) is to make a random sequence with a range of values from 1 to 900, but as many as 800 values are taken.

```r
set.seed(100)
indeks_training_set <-sample(900,800)

input_training_set <-data.feed[indeks_training_set,]
class_training_set <-data[indeks_training_set,]$risk_rating
input_testing_set <- data.feed[-indeks_training_set,]
```

Syntax usage of the C5.0 function is as follows:
<pre>
    C5.0 (input_variables, class_variables)

```r
model <- C5.0(data.feed, data$risk_rating, control = C5.0Control(label="Risk Rating"))

summary(model)
```

### Predict with Model

<pre>
    predict(model, test_set)
</pre>  

```r
predict(model, input_testing_set)
```

### Combining Prediction Results
As informed in the previous section, we will store risk_rating from the initial dataset and the results of this prediction into two other column names in the input_testing_set data frame. Let's name the column risk_rating and hasil_prediksi.

We use the first command to save the original risk_rating value into the risk_rating column.
<pre>
    input_testing_set$risk_rating <- data[-index_training_set,]$risk_rating
</pre>

Note: Remember if -index_training_set (with a minus sign in front) shows index numbers for the testing set.

Then use the second command to save the predicted value in the hasil_prediksi column.

<pre>
    input_testing_set$hasil_prediksi  <- predict (risk_rating_model, input_testing_set)
</pre>


```r
input_testing_set$risk_rating <- data[-indeks_training_set,]$risk_rating
input_testing_set$hasil_prediksi <- predict(model, input_testing_set)
head(input_testing_set, 5)
```

### Make a Confusion Matrix

To produce this we can use the dcast function from the reshape2 package, the method is as follows.
<pre>
    dcast(kolom ~ baris, dataframe)
